# Limpieza

```python
cwd = os.getcwd()
# datos scrapeados del sitio Properati
data = pd.read_csv(cwd + '\\departamentos_venta_2022_properati.csv', index_col=None, header=0)
```

```python
data.head(15)
```

```python
lista_series = data['0']
lista_series[0:15]
```

```python
# hay 4469 precios 
print(lista_series[lista_series.str.contains('USD') == True])
```

```python
# 2961 tienen el atributo ambiente o ambientes
print(lista_series[lista_series.str.contains('ambiente') == True])
```

```python
# 4410 tienen el atributo baño o baños
print(lista_series[lista_series.str.contains('baño') == True])
```

```python
# dejo registros de precio y m2    
lista_series.dropna(inplace=True)  
lista_series = lista_series[lista_series.str.contains('Publicado\nhace') == False]
lista_series = lista_series[lista_series.str.contains('Expensas') == False]
lista_series = lista_series[lista_series.str.contains('baño') == False]
lista_series = lista_series[lista_series.str.contains('ambiente') == False]
```

```python
lista_series.reset_index(drop=True,inplace=True)
lista_series[0] = '|' + lista_series[0]   # modifico primer valor para no perderlo luego en la transformacion 
lista_series
```

```python
lista_string = lista_series.str.cat(sep='|') # formo un solo string
# separo por lo que divide a cada publicacion 
lista_str_separado = lista_string.split('Publicado') 
lista_str_separado[0:10]
```

```python
# hay un espacio en blanco al final
print(len(lista_str_separado))
lista_str_separado.remove('')
print(len(lista_str_separado))
```

```python
serie = pd.Series(lista_str_separado)
df = serie.str.split('|',expand=True)
df.head()
```

```python
df.info()
```

```python
df.drop(columns=[0,5,6], inplace=True)
df.head()     
```

```python
# hay 1046 departamentos duplicados
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
# elimino columna Inmobiliaria
df.drop(columns=[4],axis=1,inplace=True)
```

```python
# hay 2 valores en columna de precios incorrecto
print(len(df[df[1].str.contains('USD') == False]))
# 1 valor incorrecto
print(len(df[df[2].str.contains('Capital Federal') == False]))
# 14 valores incorrectos en m2 
print(len(df[df[3].str.contains('m²') == False]))
```

```python
df = df[df[1].str.contains('USD') == True]
df = df[df[2].str.contains('Capital Federal') == True]
df = df[df[3].str.contains('m²') == True]
df.reset_index(drop=True,inplace=True)
# renombrar columnas
df.rename(columns={1:'Precio(USD)',2:'Barrio',3:'M2'},inplace=True)
df['Precio(USD)'] = df['Precio(USD)'].str.split('USD',expand=True)[1].replace({' ':'','\.':''},regex=True)
df.head()
```

```python
# separar metros cuadrados cubiertos y metros cuadrados totales
```

```python
# hay 1352 registros con solamente info metros cuadrados totales (se asume que es igual a la cantidad de metros cuadrados cubiertos)
df[df['M2'].str.contains('/') == False]
```

```python
df_metros_cuadrados = df['M2'].str.split('/',expand=True)
df_metros_cuadrados.info()
```

```python
df_metros_cuadrados[df_metros_cuadrados[1].isnull() == True]
```

```python
indices_valores_nulos = df_metros_cuadrados[df_metros_cuadrados[1].isnull() == True].index

for i in indices_valores_nulos:
    valor_a_remplazar = df_metros_cuadrados.at[i,0]
    df_metros_cuadrados.at[i,1] = valor_a_remplazar
```

```python
df_metros_cuadrados.info()
```

```python
df_modif = pd.concat([df,df_metros_cuadrados],axis=1)
df_modif.rename(columns={0:'m2_cubiertos',1:'m2_totales'},inplace=True)
df_modif.head()
```

```python
df_modif.tail()
```

```python
df_modif.drop(columns=['M2'],inplace=True)
df_modif.head()
```

```python
df_modif['m2_cubiertos'] = df_modif['m2_cubiertos'].replace({'m²':'',' ':''},regex=True)
df_modif['m2_totales'] = df_modif['m2_totales'].replace({'m²':'',' ':''},regex=True)
df_modif.head()
```

```python
df_modif.tail()
```

```python
df_barrio = df_modif['Barrio'].str.split(', ',expand=True)
df_barrio
```

```python
# en Barrio hay desagregaciones en Palermo
df_modif[df_modif.Barrio.str.contains(', Palermo') == True] 
```

```python
df_barrio[df_barrio[1] == 'Palermo']
```

```python
# copio Palermo en primer columna

indices = df_barrio[df_barrio[1] == 'Palermo'].index

for i in indices:
    valor_a_remplazar = df_barrio.at[i,1]
    df_barrio.at[i,0] = valor_a_remplazar
```

```python
df_barrio[df_barrio[1] == 'Palermo']
```

```python
barrios = df_barrio[0]
df_modified = pd.concat([df_modif,barrios],axis=1)
df_modified.rename(columns={0:'BARRIO'},inplace=True)
df_modified
```

```python
df_modified.drop(columns=['Barrio'],inplace=True)
```

```python
df_modified.head()
```

```python
print(df_modified.BARRIO.unique())
```

```python
df_modified
```

```python
# Transformaciones tipo de dato
df_modified['Precio(USD)'] = df_modified['Precio(USD)'].astype('int64')
df_modified['m2_cubiertos'] = df_modified['m2_cubiertos'].astype('int64')
df_modified['m2_totales'] = df_modified['m2_totales'].astype('int64')

# creo columna dolares por metro cuadrado
df_modified['USD_m2'] = (df_modified['Precio(USD)']/df_modified['m2_cubiertos']).astype('int64')
```

```python
df_modified.reset_index(drop=True,inplace=True)
```

```python
df_modified.info()
```

```python
#  quitar tildes de Barrio
a,b = 'áéíóú','aeiou' 
trans = str.maketrans(a,b) 
df_modified['BARRIO'] = df_modified['BARRIO'].str.translate(trans)
# Barrio en mayusculas
df_modified['BARRIO'] = df_modified['BARRIO'].str.upper()
```
